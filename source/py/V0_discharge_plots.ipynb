```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plot_variables(file_name, title, output_file):
    # Set Seaborn style and color palette
    sns.set(style="darkgrid")
    sns.set_palette("colorblind")

    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(file_name)

    # Convert the "datetime" column to datetime type
    df['datetime'] = pd.to_datetime(df['datetime'])


    # Extract the variables of interest
    pcr_dis = df['pcr_dis']
    wg3_dis = df['wg3_dis']
    lis_dis = df['lis_dis']
    obs = df['obs']

    # Create a figure and axes for the subplots
    fig, axes = plt.subplots(nrows=4, ncols=1, figsize=(10, 8))
    fig.suptitle(title)

    # Plot the variables in the corresponding subplots
    axes[0].plot(df['datetime'], pcr_dis, linestyle='-', color='black')
    axes[0].set_ylabel('PCRGLOBWB')
    axes[1].plot(df['datetime'], wg3_dis, linestyle='--', color='black')
    axes[1].set_ylabel('Watergap3')
    axes[2].plot(df['datetime'], lis_dis, linestyle='-.', color='black')
    axes[2].set_ylabel('LISFLOOD')
    axes[3].plot(df['datetime'], obs, linestyle=':', color='black')
    axes[3].set_ylabel('Observed')


    # Set common labels for the x and y axes
    fig.text(0.5, 0.04, 'Time', ha='center')
    #fig.text(0.04, 0.5, 'Flow Depth (m/day)', va='center', rotation='vertical')

    # Adjust the spacing between subplots
    plt.subplots_adjust(hspace=0.4)

    # Save the figure
    plt.savefig(output_file)

    # Display the plot
    plt.show()
```

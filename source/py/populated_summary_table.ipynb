```python
import os
import pandas as pd
```

```python
directory = '../R/viz/'
```

```python
os.chdir(directory)
```

```python
data_dict_columns = {
    "average": "Ensemble Mean",
    "all_dis": "All Discharge",
    "all_dis_meteo": "All Discharge Meteo",
    "allpredictors": "All Predictors"
}


setup = ["all_stations", "rhine_only", "rhine_elbe", "rhine_maas"]
model = "MLR"
basename = "cumulative_kge_"
combined_df = pd.DataFrame()  # Initialize an empty DataFrame for the combined data

for i in range(len(setup)):
    file_name = f"{basename}{setup[i]}_{model}.csv"
    print(file_name)
    
    df = pd.read_csv(os.path.join(setup[i], file_name))

    
   # Select rows where the "setup" column matches the keys in data_dict_columns
    selected_rows = df[df["setup"].isin(data_dict_columns.keys())]
    
# Replace the values in the "setup" column with the corresponding values from data_dict_columns
    selected_rows["setup"].replace(data_dict_columns, inplace=True)
    # Add a new column "folder" and populate it with setup[i]
    selected_rows.loc[:, "folder"] = setup[i]
    
    # Append the selected and renamed DataFrame to the combined DataFrame
    combined_df = pd.concat([combined_df, selected_rows], ignore_index=True)

```

```python
combined_df = combined_df[combined_df["setup"].isin(data_dict_columns.values())]
combined_df
```

```python
combined_df.to_csv(f"cumulative_kge_{model}.csv")
```
